```python
# Read Gaia data and compute aux quantities
gaia_data = pd.read_csv("/home/dmichalik/ESTEC/Home/PROJECT2 - Hyades/dr2code/trunk/all_stars_in_100pc_filtered_SNR10.csv")
gaia_data['abs_mag'] = gaia_data['phot_g_mean_mag'] + 5 * np.log10(gaia_data['parallax']/100)
gaia_data['uwe'] = np.sqrt(gaia_data['astrometric_chi2_al'] / (gaia_data['astrometric_n_good_obs_al'] - 5))

ruwe = []
for rdx,row in gaia_data.iterrows():
    g_mag_rounded = (int(row['phot_g_mean_mag'] * 100) / 100.0)
    bp_rp_rounded = (int(row['bp_rp'] * 10) / 10.0)
    u0val = u0[u0['g_mag'] == g_mag_rounded and u0[' bp_rp'] == bp_rp_rounded][' u0']
    ruwe.append(row['uwe'] / u0val)

gaia_data['ruwe'] = ruwe

plt.figure()
plt.hist(gaia_data['ruwe'], 100)

# Plot CMD
plt.figure()
plt.gca().invert_yaxis()
plt.scatter(gaia_data['bp_rp'], gaia_data['abs_mag'],s=3,alpha=0.01)
```
