```python
import numpy as np
import pandas as pd
from astropy.table import Table, vstack
from matplotlib import pyplot as plt
from glob import glob
from astropy.stats import LombScargle
from astroquery.simbad import Simbad
import astropy.coordinates as coord
import astropy.units as u
from scipy import stats
from astropy.table import Table
import warnings
#from TESStools import *

# ruwe = renormalized unit weight error
# For explanations see https://www.cosmos.esa.int/web/gaia/dr2-known-issues
# ruwe tools implementation from https://github.com/agabrown/gaiadr2-ruwe-tools with updated folder structure
from ruwe.ruwetools import U0Interpolator
rwi = U0Interpolator('ruwe/table_u0_g_col.txt', 'ruwe/table_u0_g.txt')

%matplotlib inline
```

```python
short_cadence_xmatch = '../../data/tess_gaia_magnitudes.csv'

data = pd.read_csv(short_cadence_xmatch)
```

```python
len(data)
```

```python
#Compute auxiliary columns for quality filtering
data['uwe'] = np.sqrt(data['astrometric_chi2_al'] / (data['astrometric_n_good_obs_al'] - 5))
data['ruwe'] = data['uwe'] / rwi.get_u0(data['phot_g_mean_mag'], data['bp_rp']) 

# astrometric filter on renormalized unit weight error
ruwefilter = data['ruwe'] < 1.40 # see histogram of ruwe values to verify cutoff point - see IAU 2018 presentation by Lindegren on https://www.cosmos.esa.int/web/gaia/dr2-known-issues

# photometry quality filter, see Gaia DR2 release paper by Lindegren et al. 2018
# 1.0 + 0.015(G_{BP} − G_{RP} )^2 < phot_bp_rp_excess_factor < 1.3 + 0.06(G_{BP} − G_{RP})^2
photometryfilter = ((1.0 + 0.015 * (data['bp_rp'])**2 < data['phot_bp_rp_excess_factor']) & 
                    (1.3 + 0.06  * (data['bp_rp'])**2 > data['phot_bp_rp_excess_factor']))

len(data[ruwefilter][photometryfilter])

plt.hist(data['ruwe'], 1000)
plt.xscale('log')
plt.axvline(1.4, color='r')
```

```python
#accounting for nonexistant A_G
nanmags = data[np.isnan(data['M_G'])]
nanmags['M_G_low'] = nanmags['phot_g_mean_mag'] - 5.0*np.log10(nanmags['r_est']) + 5
nanmags['BR_hi'] = nanmags['phot_bp_mean_mag'] - nanmags['phot_rp_mean_mag']
```

```python
plt.hexbin(data['ra'],data['dec'],mincnt=1,bins='log')
plt.xlim(0,360)
plt.ylim(-90,90)
```

```python
#2018A&A...616A..12G (Helmi et al. 2018) did the filtering for stars in the MCs
lmc_ids = Table.read('lmc_gaiaids.fits',format='fits').to_pandas()
smc_ids = Table.read('smc_gaiaids.fits',format='fits').to_pandas()
```

```python
lmc_data = pd.merge(data,lmc_ids,left_on='source_id',right_on='Source')
smc_data = pd.merge(data,smc_ids,left_on='source_id',right_on='Source')
mcs = pd.concat([lmc_data,smc_data])
galactic_data = data[~data['source_id'].isin(mcs['source_id'])]

#Coefficients from Malhan, Ibata & Martin (2018a), ratio of A_X/A_V
ag_av = 0.85926
ab_av = 1.06794
ar_av = 0.65199

lmc_RV = 3.41 #Gordon et al. 2003
lmc_EBV = 0.13 #Massey et al. 2007
lmc_EBPRP = (lmc_RV*lmc_EBV)*(ab_av-ar_av)
lmc_AG = (lmc_RV*lmc_EBV)*ag_av
lmc_data['(BP_RP)_0'] = lmc_data['phot_bp_mean_mag'] - lmc_data['phot_rp_mean_mag'] - lmc_EBPRP
#distance modulus from Kovacs 2000b
lmc_data['M_G'] = lmc_data['phot_g_mean_mag'] - 18.52 - lmc_AG #kovacs 2000

smc_RV = 2.75 #Gordon et al. 2003
smc_EBV = 0.09 #Massey et al. 2007
smc_EBPRP = (smc_RV*smc_EBV)*(ab_av-ar_av)
smc_AG = (smc_RV*smc_EBV)*ag_av
smc_data['(BP_RP)_0'] = smc_data['phot_bp_mean_mag'] - smc_data['phot_rp_mean_mag'] - smc_EBPRP
#distance modulus from Kovacs 2000a
smc_data['M_G'] = smc_data['phot_g_mean_mag'] - 19.05 - smc_AG #kovacs 2000
```
